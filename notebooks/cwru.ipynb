```python
import pandas as pd
from pathlib import Path

log_path = Path("/root/autodl-tmp/metro-fault-diagnosis/logs/cwru_train_log.csv")
df = pd.read_csv(log_path)
df.head()
```

```python
import matplotlib.pyplot as plt

epochs = df["epoch"].values

plt.figure()
plt.plot(epochs, df["train_loss"].values)
plt.xlabel("Epoch")
plt.ylabel("Train Loss")
plt.title("Training Loss")
plt.show()

plt.figure()
plt.plot(epochs, df["val_acc"].values, label="val_acc")
plt.plot(epochs, df["val_bal"].values, label="val_bal_acc")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Validation Accuracy / Balanced Accuracy")
plt.legend()
plt.show()
```

```python
best_i = df["val_bal"].values.argmax()
best_epoch = int(df.loc[best_i, "epoch"])
best_bal = float(df.loc[best_i, "val_bal"])

print("Best epoch:", best_epoch, "best val_bal:", best_bal)

import matplotlib.pyplot as plt
plt.figure()
plt.plot(epochs, df["val_bal"].values)
plt.axvline(best_epoch, linestyle="--")
plt.xlabel("Epoch")
plt.ylabel("val_bal_acc")
plt.title("Validation Balanced Accuracy (best marked)")
plt.show()
```
